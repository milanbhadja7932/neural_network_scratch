```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.ylabel('length')
plt.xlabel('height')
car_height = [4,3.5,3.5,4,4.2,3.9,4.5]
truck_height= [1,2,2,2.5,2.6,2.9,2.3]
car_length=[5,6,7,8,7.5,6.6,7.9]
truck_length=[10,12,13,14,12.5,12.9,13.2]
plt.plot(car_height,car_length,'ro',truck_height,truck_length,'bo')
plt.axis([0,10,0,20])
plt.show()
```

RANDOM SEPERATE LINE


```
plt.plot(car_height,car_length,'ro',truck_height,truck_length,'bo',[2,4,5,6,7],[2,6,8,10,12])
plt.ylabel('length')
plt.xlabel('height')
plt.show()
```

Random SEPARATE LINE

```
plt.plot(car_height,car_length,'ro',truck_height,truck_length,'bo',[1.4,1.5,1.6,1.7,1.8],[6,8,10,12,14])
plt.ylabel('length')
plt.xlabel('height')
plt.show()
```

PERFECT SEPERATE LINE

```
plt.plot(car_height,car_length,'ro',truck_height,truck_length,'bo',[1,2,3,4],[6,8,10,12,])
plt.ylabel('length')
plt.xlabel('height')
plt.show()
```

UNKNOWN DATA


```
plt.plot(car_height[1]-1,car_length[1]+2,'go',[1,2,3,4],[6,8,10,12,])
plt.ylabel('length')
plt.xlabel('height')
plt.axis([1,4.5,6,14])
plt.show()
```

GRAPH FOR ONLY ONE EXAMPLE


```
car_height_one=3
car_length_one=1
truck_height_one=1
truck_length_one=3
plt.plot(car_height_one,car_length_one,'ro',truck_height_one,truck_length_one,'bo')
plt.ylabel('length')
plt.xlabel('height')
plt.axis([0,5,0,5])
plt.show()
```

```
x=[1,2,3,4]
y=[]
for i in x:
    y.append(x[i-1]*0.25)
plt.plot(car_height_one,car_length_one,'ro',truck_height_one,truck_length_one,'bo',x,y)
plt.ylabel('length')
plt.xlabel('height')
plt.show()
```

```
x_real=[1,2,3,4]
y_real=[]
for i in x_real:
    y_real.append(x_real[i-1]*0.35)
plt.plot(car_height_one,car_length_one,'ro',truck_height_one,truck_length_one,'bo',x,y,'b-',x_real,y_real,'g-')
plt.ylabel('length')
plt.xlabel('height')
plt.text(2.25, 0.3, r'calculated random line',fontsize=14, color='red')
plt.text(2.5, 3,r'perfect line',fontsize=14, color='red')
plt.annotate('Error', xy=(3.8, 0.9), xytext=(3.25, 1.3),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )
plt.show()
```

```
x_real_dash=[1,2,3,4]
y_real_dash=[]
for i in x_real:
    y_real_dash.append(x_real_dash[i-1]*2.90)
plt.plot(car_height_one,car_length_one,'ro',truck_height_one,truck_length_one,'bo',x,y,'b-',x_real,y_real,'g-',x_real_dash,y_real_dash,'y-')
plt.text(1.25, 4.3, r'2.90  line',fontsize=14, color='red')
plt.ylabel('length')
plt.xlabel('height')
plt.show()
```

```
x_real_dash_final=[1,2,3,4]
y_real_dash_final=[]
for i in x_real_dash_final:
    y_real_dash_final.append(x_real_dash_final[i-1]*1.6)
plt.plot(car_height_one,car_length_one,'ro',truck_height_one,truck_length_one,'bo',x,y,'b-',x_real,y_real,'g-',x_real_dash,y_real_dash,'y-',x_real_dash_final,y_real_dash_final,'r-')
plt.ylabel('length')
plt.xlabel('height')
plt.text(1.75, 2.3, r'Final line',fontsize=14, color='red')
plt.show()
```
